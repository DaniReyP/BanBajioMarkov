Por lo que en este estudio suponemos que es constante, es decir que no existe en ese momento ningún fenómeno que haga que cambie de forma drástica el volumen de acciones en el mercado.

### Precio 
##### El precio de las acciones puede depender de diversas causas, entre ellas se encuentran la rentabilidad de las acciones, es decir el rendimiento que generan, la volatilidad del mercado, los principios de oferta y demanda, los resultados financieros que obtenga la empresa emisora, la inflación, el ciclo económico del país, etc.

### Proceso de Markov 
##### En nuestro proceso usaremos el modelo de Proceso de la cadena de Markov, que predice las tendencias de las acciones y los mercados de valores, por lo que se considera una herramienta muy útil para el análisis técnico. 
##### El principio básico del modelo de cadena de Markov es, que predice el estado de un objeto en cierto periodo de tiempo en el futuro con un vector de probabilidad del estado inicial y la matriz de probabilidad del estado de transición. 
##### Una diferencia entre este método de pronóstico y los métodos estadísticos como la regresión, series de tiempo, ARIMA, etc., son que la cadena de Markov no tiene que encontrar características mutuas entre los factores del predictor para tener en cuenta las características de la evolución histórica. 
#### La definición de cadena de Markov 
##### Deje que un espacio discreto de secuencia aleatoria $E$. Si por cualquier número entero no negativo $n1,n2,...,nm(0≤n1<n2<...<nm)$ y un número natural arbitrario $k$, así como la arbitrariedad $i1, i2, ..., im, j\ \epsilon\ E$, satisfaciendo la siguiente condición: 

$ P\{X_{nm+i}=j| X_({n_1})=i_{1},  X_({n_2})=i_{2}, ...,  X_({nm})=i_{m}\} = P\{X_{nm+i}=j| X_({km})=i_{m} \}$

##### Entonces esta secuencia aleatoria $\{X_{(n)}, n = 1,2,...\}$ se dice que es la cadena de Markov.
### Características del Modelo 
##### Suposiciones en nuestro modelo de la cadena de Markov para la predicción de la tendencia del precio de la acción de BanBajío.

##### El precio de una acción se ve afectado por factores aleatorios de carácter macroeconómico local y mundial, político, social, etc. 
##### El alza o la baja del precio de la acción en un día determinado sólo dependía del estado inmediato anterior del precio de cierre, el pasado es insignificante. 
##### La probabilidad de que el precio de la acción de un estado i se salta a otro estado j en el mismo intervalo de tiempo no tiene nada que ver con el momento del estado i.

#### Pasos para pronosticar:
#### La construcción del proceso del estado y la determinación de la probabilidad del estado. 
##### Si tomamos cada día de cierre como unidades de tiempo discreto, los precios de cierre se dividen en tres estados: alza, mantenimiento(cero), baja. Y dejamos $x_{1}=$ alza, $x_{2}=$  mantenimiento(cero), $x_{3}=$  baja, entonces el espacio de estados es $E(x_{1},x_{2},x_{3})$, y la probabilidad del estado es el tamaño de la posibilidad de aparición de una variedad de estado. El vector de estado se denota por $n_{(i)} = (p_{1},p_{2}, ...,p_{n})$ donde $i=1,\ 2,\ ...,\ n,\ p,$ es la probabilidad de $x,\ j\ =\ 1,\ 2,\ ...,\ n$.

##### Y el vector de estado inicial,  $n_{(0)} = (p_{1},p_{2},p_{3}) =(.5789,.05263,.3684)$
#### El establecimiento del estado de la matriz de probabilidad de transición.
##### Dado que el estado del último día es hacia la baja mientras que no hay una transición de estado, el número total del primer estado debe registrarse como $n-1$, es decir $7-1=6$ veces.
1. La probabilidad de alza a alza es $p_{11}=\ \frac{7}{11}$, alza a cero es $p_{12}=\ \frac{0}{11}$ y alza a baja $p_{13}=\ \frac{4}{11}$

##### Expresamos cada probabilidad de transición de estado mediante la matriz de transición de estado cerrado.
##### $P=\ \left[\begin{matrix}P_{11}&P_{12}&P_{13}\\P_{21}&P_{22}&P_{23}\\P_{31}&P_{32}&P_{33}\\\end{matrix}\right]=$ $\ \left[\begin{matrix}.6364&0&.3636\\0&0&1\\.6667&.1667&.1667\\\end{matrix}\right]$

##### Donde cada fila de la matriz $P$ es la probabilidad de transición de estado de diversas situaciones.
#### Cálculo de la probabilidad de estado de los días de cierre subsiguientes por la matriz de probabilidades de transición. 
##### De acuerdo con el proceso de Markov, la probabilidad del estado en diferentes periodos se indica mediante $n_{(i)}, n_{(i+1)}=n_{(i)}\ast\ P$, donde recordemos que $P$ es la matriz de transición de estado. La información del precio de la acción de BanBajío es de 22 días ya que la BMV opera únicamente de lunes a viernes. Tomamos como vector de estado inicial $n_{(0)}=(1,0,0)$, con lo que buscamos predecir la probabilidad de estado de diversas fechas de cierre en el futuro, por lo que obtendremos el vector de probabilidad de precio de cierre del día 23:
$n_{(1)}=n_{(0)}\ast\ P=\ \left[0\ 0\ 1\right]\left[\begin{matrix}.6364&0&.3636\\0&0&1\\.6667&.1667&.1667\\\end{matrix}\right]=(.6667,.1667,.1667)$

##### Por lo tanto, el vector de probabilidad de estado de precio de cierre del día 21 es $n_{(2)}=n_{(1)}\ast\ P$.

#### Estados Recurrentes

##### Un conjunto de estados recurrentes forman una cadena sencilla si cada uno de los estados del conjunto se comunica con cada uno de los demás estados del conjunto

##### En este caso se presenta una cadena sencilla ya que un estado $j$ es recurrente (o persistente) si con probabilidad 1, el proceso eventualmente regresará al estado $j$ después de haber salido de él, y podemos deducir que nuestros 3 estados presentan esta característica de recurrencia.

```python
#Precio de la acción BBAJIOO de octubre del 2017 a octubre del 2020 

import pandas_datareader as pdr
import matplotlib.pyplot as plt
import datetime

inbursa =pdr.get_data_yahoo('BBAJIOO.MX', start=datetime.datetime(2017,10,12), end=datetime.datetime(2020,10,13))
column = inbursa['Adj Close']

column.plot(grid=True)
plt.show()
```

```python
import numpy as np
import random as rn

estados = ["Alza", "Cero", "Baja"]

matrizEstados = [["P11", "P12", "P13"], ["P21", "P22", "P23"], ["P31", "P32", "P33"]]

matrizTransicion =[[0.6364, 0, 0.3636], [0, 0, 1], [0.6667, 0.1667, 0.1667]]

def prediccion (dias):
    accionHoy = "Alza"
    print("Estado inicial " + accionHoy)
    accionVector = [accionHoy]
    i = 0
    prob = 1
    while i != dias:
      if accionHoy == "Alza":
        variacion = np.random.choice(matrizEstados[0], replace=True, p=matrizTransicion[0])
        if variacion == "P11":
          prob = prob * 0.6364
          accionVector.append("Alza")
          pass
        elif variacion == "P12":
          prob = prob * 0
          accionHoy = "Cero"
          accionVector.append("Cero")
        else:
          prob = prob * 0.3636
          accionHoy = "Baja"
          accionVector.append("Baja")
      elif accionHoy ==  "Cero":
        variacion = np.random.choice(matrizEstados[1],replace=True,p=matrizTransicion[1])
        if variacion == "P21":
          prob = prob * 0
          accionVector.append("Alza")
          pass
        elif variacion == "P23":
          prob = prob * 0
          accionHoy = "Cero"
          accionVector.append("Cero")
        else:
          prob = prob * 1
          accionHoy = "Baja"
          accionVector.append("Baja")
      elif accionHoy ==  "Baja":
        variacion = np.random.choice(matrizEstados[2],replace=True,p=matrizTransicion[2])
        if variacion == "P31":
          prob = prob * 0
          accionVector.append("Alza")
          pass
        elif variacion == "P32":
          prob = prob * 0.5556
          accionHoy = "Cero"
          accionVector.append("Cero")
        else :
          prob = prob * 0.4444
          accionHoy = "Baja"
          accionVector.append("Baja")

      i += 1

    print("Ultimo estado despues de " + str(dias) + " dias: " + accionHoy)
    print("Probabilidad de la secuencia de estados " + str(prob))
    print("Posibles estados: " + str(accionVector))

prediccion(1)
```
